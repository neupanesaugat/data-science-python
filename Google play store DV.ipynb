```python
df.columns = df.columns.str.replace(" ","_")
df.columns
```

HANDLING MISSING DATA

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
sns.set(rc = {"figure.dpi":300, "figure.figsize":(12,9)})
```

```python
sns.heatmap(df.isnull(), cbar=False)
```

```python
#Median
rating_median = df["Rating"].median()
print(rating_median)
df["Rating"].fillna(rating_median, inplace = True)
```

```python
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

DATA PROCESSING

```python
df["Reviews"].describe()
```

```python
df["Reviews"] = df["Reviews"].astype("int64")
```

```python
df["Reviews"].describe().round()
```

```python
print(len(df["Size"].unique()))
df["Size"].unique()
```

```python
df["Size"].replace("M","", regex = True, inplace = True)
df["Size"].replace("k","", regex = True, inplace = True)
```

```python
df["Size"].unique()
```

```python
sm = df[df["Size"] !="Varies with device"]["Size"].astype(float).median()
print(sm)
```

```python
df["Size"].replace("Varies with device",sm, inplace = True)
```

```python
df["Size"].head()
```

```python
df.Size = pd.to_numeric(df.Size)
df.Size.describe().round()
```

HANDLING THE INSTALLS COLUMN

```python
df["Installs"].unique()
```

```python
df.Installs = df.Installs.apply(lambda x:x.replace("+",""))
df.Installs = df.Installs.apply(lambda x:x.replace(",",""))
df.Installs = df.Installs.apply(lambda x:int(x))
```

```python
df["Installs"].unique()
```

HANDLING THE PRICE COLUMN

```python
df["Price"].unique()
```

```python
df.Price = df.Price.apply(lambda x:x.replace("$",""))
df.Price = df.Price.apply(lambda x:float(x))
```

```python
df["Price"].unique()
```

HANDLING GENRE COLUMNS

```python
len(df["Genres"].unique())
```

```python
df["Genres"].head(10)
```

```python
df["Genres"] = df["Genres"].str.split(";").str[0]
```

```python
df["Genres"].unique()
```

```python
len(df["Genres"].unique())
```

```python
df["Genres"].str.split(";").str[1]
df["Genres"].unique()
```

```python
df["Genres"].value_counts()
```

```python
df["Genres"].replace("Music & Audio","Music", inplace= True)
```

```python
df["Genres"].unique()
```

```python
df["Last_Updated"].head()
```

```python
df["Last_Updated"]= pd.to_datetime(df["Last_Updated"])
df["Last_Updated"].head()
```
